```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
rted_res = {0: {0: {'stg_idx':'PV_1', 'bu':0.3}, 1: {'stg_idx':'PV_2', 'bu':0.3}, 2: {'stg_idx':'PV_3', 'bu':0.4}},
            1: {0: {'stg_idx':'PV_1', 'bu':0.2}, 1: {'stg_idx':'PV_2', 'bu':0.2}, 2: {'stg_idx':'PV_3', 'bu':0.6}},
            2: {0: {'stg_idx':'PV_1', 'bu':0.1}, 1: {'stg_idx':'PV_2', 'bu':0.1}, 2: {'stg_idx':'PV_3', 'bu':0.8}}}

bu = np.array(-1 * np.ones((3, 3)))
bd = np.array(-1 * np.ones((3, 3)))
for rted_idx in rted_res.keys():
    bu[rted_idx] = pd.DataFrame.from_dict(rted_res[rted_idx]).loc['bu'].values
    bd[rted_idx] = pd.DataFrame.from_dict(rted_res[rted_idx]).loc['bu'].values

bu_df = pd.DataFrame(bu, columns=['PV_1', 'PV_2', 'PV_3'])
bd_df = pd.DataFrame(bu, columns=['PV_1', 'PV_2', 'PV_3'])
```

```python
import matplotlib.font_manager
plt.style.use('ieee')

figs, axes = plt.subplots(1, 2, figsize=(10,4))
bu_df.plot.bar(stacked=True, ax=axes[0], legend=False)
bd_df.plot.bar(stacked=True, ax=axes[1], legend=False)

for ax in axes:
    ax.set_xticks(ticks=axes[0].get_xticks(), labels=['PV_1', 'PV_2', 'PV_3'])
    ax.tick_params(axis='x', labelrotation = 0)
    ax.set_ylim([0, 1])
```
