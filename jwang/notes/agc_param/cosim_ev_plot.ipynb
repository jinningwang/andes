```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
os.chdir('../')
```

```python
dir_path = os.path.abspath('..')
case1 = pd.read_csv(dir_path + '/notes/test_ev1.csv')
case2 = pd.read_csv(dir_path + '/notes/test_ev2.csv')
```

```python
plt.style.use('ieee')
fig, ax = plt.subplots(1, 3, figsize=(15, 4))
case1.plot(x='time', y='ace', ax=ax[0], label='Case1')
case2.plot(x='time', y='ace', ax=ax[0], label='Case2')
case1.plot(x='time', y='freq', ax=ax[1], label='Case1')
case2.plot(x='time', y='freq', ax=ax[1], label='Case2')
case1[['fdev']].plot(kind='kde', ax=ax[2], label='Case1')
case2[['fdev']].plot(kind='kde', ax=ax[2], label='Case2')
ax[0].set_xlim(case1['time'].iloc[0], case1['time'].iloc[-1])
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('ACE [MW]')
ace1 = case1['ace'].std()
ace2 = case2['ace'].std()
ax[0].set_title(f'$\sigma1$={ace1.round(4)}, $\sigma2$={ace2.round(4)}')
ax[1].set_xlim(case1['time'].iloc[0], case1['time'].iloc[-1])
ax[1].set_xlabel('Time [s]')
ax[1].set_ylabel('COI Freq. [Hz]')
ax[2].set_xlabel('Freq. Dev. [Hz]')
ax[2].legend(['Case1', 'Case2'])
fd1 = case1['fdev'].std()
fd2 = case2['fdev'].std()
ax[2].set_title(f'$\sigma1$={fd1.round(4)}, $\sigma2$={fd2.round(4)}')
```
