```python
reg_data['reg_hourly'] = reg_data['rega_hourly'] + reg_data['regd_hourly']
plt.style.use('ieee')
fig, ax = plt.subplots(figsize=(5,3))
for i in range(30):
    reg_data.iloc[24*i:24*(i+1)].plot(x='datetime_beginning_ept', y='reg_hourly', ax=ax, label=f'Day{i+1}')
ax.set_xticks([i for i in range(24)])
ax.set_xticklabels([i for i in range(24)])
ax.set_xlabel('Time [H]')
ax.set_ylabel('Reg. Mileage [MW]')
ax.set_xlim(0, 23)
ax.legend('')
ax.set_title('PJM RTO Mileage in June')
```

```python
plt.style.use('ieee')
fig, ax = plt.subplots(figsize=(5,3))
for i in range(30):
    rev_data[rev_data['service']=='REG'].iloc[24*i:24*(i+1)].plot(x='datetime_beginning_ept', y='as_mw', ax=ax, label=f'Day{i+1}')
ax.set_xticks([i for i in range(24)])
ax.set_xticklabels([i for i in range(24)])
ax.set_xlabel('Time [H]')
ax.set_ylabel('Reg. Capacity [MW]')
ax.set_xlim(0, 23)
ax.legend('')
ax.set_title('PJM RTO Reg. Capacity in June')
```
