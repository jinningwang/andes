```python
class TranssmisionCenter():
    """Transmission system"""
    def __init__(self) -> None:
        pass
    def dispatch(self):
        """Do dispatch"""
        pass
    def run(self):
        """Run Dynamic Simulation"""
        pass

class DistributionCenter():
    """Distribution system"""
    def __init__(self) -> None:
        pass
    def run(self):
        """Run simulation"""
        pass

class BaseDG():
    """Base class for Distributed Generation"""

    def __init__(self) -> None:
        pass
        self.param_types = {}
        self.parm = {}

    def setup(self):
        """Setup"""
        pass

    def load(self, file_path=''):
        """Load parameters from Excel file"""
        df = pd.read_excel(file_path, encoding='utf-16')
        self.param = df.to_dict()
        return True

    def send(self):
        """Send info"""
        pass

    def receive(self):
        """Receive info"""
        pass

class DGCenter(BaseDG):
    """Base class for Distributed Generation center"""
    def __init__(self) -> None:
        super().__init__()

class EVCenter(DGCenter):
    """EV control center, based on state space model"""
    def __init__(self) -> None:
        super().__init__()

    def setup(self):
        """Setup State Space Model"""
        pass

    def efrc(self):
        """Estimate frequency regulation capacity"""
        pass

    def gctrl(self):
        """Generte control signal"""
        pass

class DGStation(BaseDG):
    """Base class for Distributed Generation Station"""
    def __init__(self) -> None:
        super().__init__()

    def run(self):
        """Run Monte-Carlo simulation"""
        pass

class EVStation(DGStation):
    """EV Station"""
    def __init__(self) -> None:
        super().__init__()

    def setup(self):
        """Setup EV devices"""
        pass

    def rctrl(self):
        """Response to control signal"""
        pass

    def run(self):
        """Run simulation"""
        pass
```

```python
# --- Initialization ---
tc = TranssmisionCenter()  # Transmission system
fd = DistributionCenter()  # Distribution feeder
evs1 = EVStation()  # EV Station
evs2 = EVStation()
evc = EVCenter()  # EV Center

# --- Dispatch ---
evc.efrc()  # EVS estimate FRC
evc.send()  # EVS send info to transmisstion
tc.dispatch()  # dispatch

# --- DG Control ---
evc.receive()  # receive info from transmission
evc.gctrl()  # generate control signal
evc.send()  # send control signal to EVStation

evs1.receive()  # receive control signal
evs1.rctrl()  # response to control signal

evs2.receive()  # receive control signal
evs2.rctrl()  # response to control signal

# --- Run ---
# Run Distribution
fd.run()

# RUN DG
evs1.run()
evs2.run()
# Federate Power with Transmission

# RUN Dynamic
tc.run()
```
